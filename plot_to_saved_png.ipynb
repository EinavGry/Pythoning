```python
from matplotlib import pyplot as plt
import numpy as np
import os

n = 100
plt.ioff()
x = np.linspace(0,2*np.pi,n)
y = ([np.sin(i) for i in x])
plt.ylabel('sin(x)')
plt.xlabel('x')
plt.title("sin function")
plt.grid(True)
plt.plot(x, y, 'o')
file_path = os.getcwd()
save_to_path = os.path.join(file_path, "figs")
os.chdir(save_to_path)
plt.savefig('sin_wave.png' ,bbox_inches='tight')
plt.show()  
#print(os.getcwd())
print("done")
os.chdir(file_path)
```
